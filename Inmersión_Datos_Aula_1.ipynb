```python
#convierto todo a string y luego separo
inmuebles.Valor.str.split()
```

```python
#cambio el tipo de dato a dataframe
inmuebles.Valor.str.split(expand=True)

#creo una variable valor para crear una nueva columna
valor = inmuebles.Valor.str.split(expand=True)
inmuebles["Moneda"] = valor[0]
inmuebles["Precio"] = valor[1]
inmuebles.sample(3)
```

```python
#verificamos los tipos de datos de las columnas creadas
inmuebles.info()
```

```python
#quitamos los puntos de los precios y los remplazamos por nada ''
inmuebles["Precio"] = inmuebles["Precio"].str.replace('.','',regex=True)
```

```python
#imprimimos un dataframe de los dos últimas columnas para verificar los cambios
inmuebles[["Precio","Barrio"]]
```

```python
#verificamos que aún no se ha cambiado el tipo de dato
inmuebles.info()
```

```python
#se crea una columna nueva de tipo de dato float dividido entre millon para que muestre datos cortos
inmuebles["Precio_Millon"] = inmuebles.Precio.astype('float')/1000000
```

```python
#se verifica la información de la nueva columna
inmuebles.info()
```

```python
#valores estadisticos del dataset
inmuebles.describe()
```

```python
#organizar la vista de los datos para mejor lectura
pd.set_option('display.precision',2)
pd.set_option('display.float_format',lambda x: '%.2f' % x)
inmuebles.describe()
```

```python
#importamos librerias seaborn y matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

#creamos nuestro lienzo para que nos permita el espacio para colocar nuestra gráfica
plt.figure(figsize=(10,6))

#creamos el gráfico del histograma
#kde muestra la curva de densidad
#hue añade una dimensión sobre otras variables
grafica = sns.histplot(data=inmuebles, x='Precio_Millon', kde = True, hue='Tipo')
grafica.set_title('Distribución de los valores de los inmuebles en millones de pesos en Bogotá')

#establecer los limites mínimos y máximos del gráfico
plt.xlim((50,1000))
plt.ylim((0,1000))

#exportamos la gráfica
plt.savefig('/content/drive/MyDrive/Inmersión_datos_alura/Valor_inmuebles.png', format = 'png')

#muestro la gráfica
plt.show
```

```python
#tomar el precio total y el area total y dividirlos
inmuebles['Valor_m2_Millon'] = inmuebles['Precio_Millon']/inmuebles['Area']
inmuebles.head()
```

```python
#calcula el promedio de cada columna de cada barrio agrupandolos con groupby
inmuebles.groupby('Barrio').mean()
```

```python
#se suman cada dato de cada registro
datos_barrio = inmuebles.groupby('Barrio').sum()
datos_barrio
```

```python
#mejor forma de calcular el valor del m2 por cada barrio
datos_barrio['Valor_m2_Barrio'] = datos_barrio['Precio_Millon']/datos_barrio['Area']
datos_barrio
```

```python
#creo un diccionario con el valor del metro cuadrado por barrio
m2_barrio = dict(datos_barrio['Valor_m2_Barrio'])
m2_barrio
```

```python
#reemplazar el valor de m2 para cada inmueble que pertenezca al mismo barrio
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles['Valor_m2_Millon'] = inmuebles['Valor_m2_Barrio'].map(m2_barrio)
inmuebles.head(5)
```

```python
#creamos una variable para almacenar el top 10 de los barrios
top_barrios = inmuebles['Barrio'].value_counts()[:10].index
top_barrios
```

```python
#convertir barrio como una columna en lugar de un indice
datos_barrio.reset_index(inplace=True)
datos_barrio
```

```python
#consultar los top barrios que esten en la lista como dataframe 
datos_barrio.query('Barrio in @top_barrios')
```

```python
#crear gráfico
plt.figure(figsize=(10,8))
ax = sns.barplot(x='Barrio', y='Valor_m2_Barrio', data = datos_barrio.query('Barrio in @top_barrios'))
ax.tick_params(axis = 'x', rotation =45)
```

```python
#crear gráfico2
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio', y='Valor_m2_Millon', data = inmuebles.query('Barrio in @top_barrios'))
ax.tick_params(axis = 'x', rotation =45)
plt.show()
```

```python
#crear gráfico3
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio', y='Valor_m2_Millon', data = inmuebles.query('Barrio in @top_barrios & Valor_m2_Millon < 15'))
ax.tick_params(axis = 'x', rotation =45)
plt.show()
```

```python
#crear gráfico4
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio', y='Area', data = inmuebles.query('Barrio in @top_barrios & Area < 500'))
ax.tick_params(axis = 'x', rotation =45)
plt.show()
```
