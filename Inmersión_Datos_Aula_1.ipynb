```python
#tomar el precio total y el area total y dividirlos
inmuebles['Valor_m2_Millon'] = inmuebles['Precio_Millon']/inmuebles['Area']
inmuebles.head()
```

```python
#calcula el promedio de cada columna de cada barrio agrupandolos con groupby
inmuebles.groupby('Barrio').mean()
```

```python
#se suman cada dato de cada registro
datos_barrio = inmuebles.groupby('Barrio').sum()
datos_barrio
```

```python
#mejor forma de calcular el valor del m2 por cada barrio
datos_barrio['Valor_m2_Barrio'] = datos_barrio['Precio_Millon']/datos_barrio['Area']
datos_barrio
```

```python
#creo un diccionario con el valor del metro cuadrado por barrio
m2_barrio = dict(datos_barrio['Valor_m2_Barrio'])
m2_barrio
```

```python
#reemplazar el valor de m2 para cada inmueble que pertenezca al mismo barrio
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles['Valor_m2_Millon'] = inmuebles['Valor_m2_Barrio'].map(m2_barrio)
inmuebles.head(5)
```

```python
#creamos una variable para almacenar el top 10 de los barrios
top_barrios = inmuebles['Barrio'].value_counts()[:10].index
top_barrios
```

```python
#convertir barrio como una columna en lugar de un indice
datos_barrio.reset_index(inplace=True)
datos_barrio
```

```python
#consultar los top barrios que esten en la lista como dataframe 
datos_barrio.query('Barrio in @top_barrios')
```

```python
#crear gráfico
plt.figure(figsize=(10,8))
ax = sns.barplot(x='Barrio', y='Valor_m2_Barrio', data = datos_barrio.query('Barrio in @top_barrios'))
ax.tick_params(axis = 'x', rotation =45)
```

```python
#crear gráfico2
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio', y='Valor_m2_Millon', data = inmuebles.query('Barrio in @top_barrios'))
ax.tick_params(axis = 'x', rotation =45)
plt.show()
```

```python
#crear gráfico3
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio', y='Valor_m2_Millon', data = inmuebles.query('Barrio in @top_barrios & Valor_m2_Millon < 15'))
ax.tick_params(axis = 'x', rotation =45)
plt.show()
```

```python
#importamos la libreria para trabajar machine learning
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#creamos variables con los datos que creemos que afectan en el valor del inmueble
X = datos_ml[['COD_UPZ_GRUPO_x']]
y = datos_ml['Precio_Millon']

#creamos el split con las variables de entreno y de prueba. test_size=pruebe con el 75% de nuestros datos, random_state= para probar con aleatoriedad
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=99)
```

```python
#creamos una variable en donde vamos a instancias nuestra regresion de forma genérica
modelo = LinearRegression()
```

```python
#ajustamos nuestros datos. El modelo ya está aplicado, hasta aquí ya esta creado el modelo de machine learning
modelo.fit(X_train, y_train)
```

```python
#validamos el modelo, con la predicción del modelo, se compara estos datos con la dataset
y_predict_test = modelo.predict(X_test)
```

```python
#se importan las librerias para aproximar los datos con un margen de error bajo
from sklearn.metrics import mean_absolute_error, r2_score

#compara los datos nuevos con lo que se predijo. se obtiene un valor de 187 millones de margen de error
baseline_mae = mean_absolute_error(y_test, y_predict_test)

#se obtiene un valor de 0.03
baseline_r2 = r2_score(y_test, y_predict_test)

print(baseline_mae, baseline_r2)
```

```python
#se toman más variables para mejorar el margen de error
X = datos_ml[['COD_UPZ_GRUPO_x','Habitaciones','Banos','CONJUNTO_CERRADO','SALARIO_ANUAL_MI','TIENE_ESCRITURA']]

y = datos_ml["Precio_Millon"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 99)
modelo_1 = LinearRegression()
modelo_1.fit(X_train, y_train)
y_predict_test = modelo_1.predict(X_test)
y_predict_train = modelo_1.predict(X_train)
mae_test = mean_absolute_error(y_test, y_predict_test)
r2_test = r2_score(y_test, y_predict_test)
mae_train = mean_absolute_error(y_train, y_predict_train)
r2_train = r2_score(y_train, y_predict_train)
print(mae_test,r2_test)
print(mae_train,r2_train)
```
